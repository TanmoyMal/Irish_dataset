```python
df = pd.read_csv("/content/Iris.csv", usecols=['setosa', 'versicolor'])

df #read the particulat row from the data set
```

```python
df["setosa"].mean()
df["versicolor"].mean()

print("Mean of setosa:", df["setosa"].mean())
print("Mean of versicolor:", df["versicolor"].mean())

#the mean value of a particular row
```

```python
df["setosa"].median()
df["versicolor"].median()

print("Median of setosa:", df["setosa"].median())
print("Median of versicolor:", df["versicolor"].median())

#the median value of a particular row
```

```python
df["setosa"].mode()[0]
df["versicolor"].mode()[0]

print("Mode of setosa:", df["setosa"].mode()[0])
print("Mode of versicolor:", df["versicolor"].mode()[0])

#the mode value of a particular row
```

```python
sns.pairplot(data[['150', '4', 'setosa', 'versicolor', 'virginica']])
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['150', '4', 'setosa', 'versicolor', 'virginica']])
plt.title('Boxplot of Features')
plt.show()
```

```python
corr_data = data[["150", "4", "setosa", "versicolor", "virginica"]].corr()
print(corr_data)

sns.heatmap(corr_data, annot=True)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
avg_measurements = data[['setosa', 'versicolor', 'virginica']].mean()
sns.barplot(x=avg_measurements.index, y=avg_measurements.values)
plt.title('Barplot of Average Measurements by Species')
plt.ylabel('Average Measurement')
plt.show()
```

```python
sns.histplot(data['150'], bins=30, kde=False)
plt.title('Histogram')
plt.xlabel('150')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.displot(data['4'], bins=30, kde=True)
plt.title('Distribution Plot')
plt.xlabel('4')
plt.ylabel('Density')
plt.show()
```

```python
sns.scatterplot(x=data['150'], y=data['4'], hue=data['setosa'] + data['versicolor'] + data['virginica'])
plt.title('Scatterplot of 150 vs. 4')
plt.xlabel('150')
plt.ylabel('4')
plt.legend(title='Species')
plt.show()
```
